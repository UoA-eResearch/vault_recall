```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
pd.set_option('display.max_colwidth', None)
df = pd.read_csv("recall_progress.csv")
df.index = pd.to_datetime(df.pop("timestamp"))
df
```

```python
df["pct"] = df.current_size_bytes / df.actual_size_bytes * 100
for f, group in df.groupby("filepath"):
    group.pct.plot(style=".-")
```

```python
df
```

```python
recalled = len(df.groupby("filepath").filter(lambda x: any(x.pct >= 200)).filepath.unique())
n_files = len(df.filepath.unique())
print(f"{recalled}/{n_files} files recalled ({round(recalled/n_files*100, 2)}%)")
```

```python
df = df.groupby("filepath").filter(lambda x: x.current_size_bytes.nunique() > 1)
df
```

```python
recall_speeds = []
for filepath, group in df.groupby("filepath"):
    bytes_changed = group.current_size_bytes.diff()
    seconds_elapsed = group.index.to_series().diff().dt.seconds
    assert group.actual_size_bytes.nunique() == 1
    max_bps = (bytes_changed[bytes_changed>0] / seconds_elapsed).max()
    if max_bps:
        print(f"{filepath} has a max recall speed of {round(max_bps / 1000 / 1000, 2)} MB/s")
        recall_speeds.append(max_bps)
    group.current_size_bytes.plot(style=".-")
print(f"Max recall speed: {round(max(recall_speeds) / 1000 / 1000, 2)} MB/s")
```
